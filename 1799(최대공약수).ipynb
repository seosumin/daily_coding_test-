#  최대공약수

- 최대공약수란 어떤 두 개 이상의 자연수들이 있을 때, 두 수의 공통된 약수 중 가장 큰 수를 말합니다. 서로 다른 두 자연수 N, M이 주어졌을 때, 두 수의 최대공약수를 출력하는 프로그램을 만드세요.
- 입력 : 124 512
- 출력 : 4
- 입력값 : 첫째 줄에 서로 다른 두 자연수 N ,M이 공백을 두고 주어집니다. (1 <= N <= 100,000) (1 <= M <= 100,000)
- 출력값 : 두 자연수 N, M의 최대공약수를 출력합니다.

## 1) 문제 이해
- 최대공약수를 구하는 초등학교 5학년 수준의 문제를 코딩으로 설계
- 두 수를 공백으로 받고, 이 두 수에 대한 최대 공약수를 구해라

## 2) 필요한건?
- 최대공약수를 쉽게 구하는 공식?
- 가장 쉽고 간단한게 짜는 법은 n번의 숫자만큼 두 숫자를 나누고 그중 공통된 가장 높은 값을 취득

```python
# 일단 짜보자
A, B=list(map(int,input().split()))
A_list=[]
B_list=[]

# A의 약수와, B의 약수를 구한다.
for i in range(1, A):
    if A % i == 0:
        A_list.append(i)
for i in range(1, B):
    if B % i == 0:
        B_list.append(i)

# 두 개의 공통된 수를 뽑는다.
unit_list=[]
for num_a in A_list:
    for num_b in B_list:
        if num_a == num_b:
            unit_list.append(num_a)
            break
# 최대값을 출력한다.
print(max(unit_list))
```


- 코드가 길어지고 속도도 느리다.
- 간결하고 속도도 올릴 수 있을까?


```python
#파이썬 내장함수 set.intersection()
# 일단 짜보자
A, B=list(map(int,input().split()))
A_list=[]
B_list=[]

# A의 약수와, B의 약수를 구한다.
for i in range(1, A):
    if A % i == 0:
        A_list.append(i)
for i in range(1, B):
    if B % i == 0:
        B_list.append(i)

print(max(list(set(A_list).intersection(B_list))))
```


- 내장함수로 길이와 시간을 좀 줄였다.

```python
#한 큐에 공배수 구하기
A, B=list(map(int,input().split()))
for i in range(min(A, B), 0, -1):
    if A% i==0 and B%i==0:
        print(i)
        break
```

- 위의 코드는 A와 B중에 가장 작은 수를 역순으로 for문을 돌려, 두 수랑 나눴을때 0이되면 출력하는 아주 간단하면서도 빠르게 공배수를 찾아준다.

- 유클리드 호제법

- 출처 : https://codingpractices.tistory.com/34
- 두 수를 서로 상대방 수로 나누어서 공약수를 구하는 알고리즘
- 이 방법은 나머지를 이용하는 방법
- 먼저 두 수를 나눈 후 나머지를 계속 대입하는 것
- 이렇게 나누었을때 마지막 나누기의 값이 최대공약수가 된다.

```python
print(10 % 12)
print(12 % 10)
print(10 % 2)
# 나머지 연산을 반복한다!
```

```python
#유클리드 호제법
while(A!=B):
    if(A>B) : A-=B
    else : B-=A
print(A)
```
