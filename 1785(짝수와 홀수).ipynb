#  짝수와 홀수 기초

- 정수 A가 입력될 때 A가 짝수인지 홀수인지 판별하는 프로그램을 작성하세요.
- 입력 : 8
- 출력 : E
- 입력값 : 첫째 줄에 정수 A가 입력됩니다. (1 <= A <= 100,000,000)
- 출력값 : 정수 A가 짝수라면 첫째 줄에 E를 출력하고, 그렇지 않으면 O를 출력합니다.

## 1) 문제 이해
- 주어진 숫자가 짝수인지 홀수인지 확인


## 2) 필요한건?
- 파이썬으로 나누기를 할수 있는지 확인하기 위한 코딩 테스트인듯
- 파이썬의 사칙연산을 정리해보자


```python
A=int(input())
if A%2==0:
    print("E")
else:
    print("O")
```
